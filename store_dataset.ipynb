```python
x=df.select_dtypes("float64","int64").iloc[:,1:]
x
```

```python
y=df['Category']
y
```

```python
from sklearn.preprocessing import StandardScaler,MinMaxScaler

x_std=StandardScaler().fit_transform(x)
x_mnmx=MinMaxScaler().fit_transform(x)

```

```python
x_std
```

```python
x_mnmx
```

```python

from sklearn.neighbors import KNeighborsClassifier
from sklearn.model_selection import cross_val_score,StratifiedKFold,train_test_split,GridSearchCV
from sklearn.linear_model import LogisticRegression

from sklearn.metrics import accuracy_score
x_train,x_test,y_train,y_test=train_test_split(x_std,y,test_size=0.25,random_state=42)
model=KNeighborsClassifier(n_neighbors=5)
model.fit(x_train,y_train)
y_pred=model.predict(x_test)
print("accuracy_score of KNeighbors:",accuracy_score(y_test,y_pred))
```

```python
model.get_params()
param_mod={"n_jobs":[5,14,24],'n_neighbors':[20,40,90],'p':[1,2]}

grid_mod=GridSearchCV(model,param_mod,cv=StratifiedKFold(10))
grid_mod.fit(x_train,y_train)
print("the best parameters of Grid_model",grid_mod.best_params_)
print("the best score of Grid_model",grid_mod.best_score_)
```

```python
reg=LogisticRegression()
reg.fit(x_train,y_train)
y_pred=model.predict(x_test)
print("accuracy_score of Logistic Regression:",accuracy_score(y_test,y_pred))
```

```python
reg.get_params()
param_reg={"C":[1,10,100],"n_jobs":[10,20,50,100],"verbose":[0,1,2]}

grid_reg=GridSearchCV(model,param_mod,cv=StratifiedKFold(10))
grid_reg.fit(x_train,y_train)
print("the best parameters of Grid_model ",grid_reg.best_params_)
print("the best score of Grid_model ",grid_reg.best_score_)
```
