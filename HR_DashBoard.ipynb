```python
y=df["Department"]


dtclf=DecisionTreeClassifier()
x=df[["Age","Projects Completed","Productivity (%)","Satisfaction Rate (%)","Feedback Score","Salary"]]
xtr,xts,ytr,yts=train_test_split(x,y,random_state=0,test_size=0.2)
gs=GridSearchCV(dtclf,param_grid={"criterion":['entropy','gini'],"max_leaf_nodes":[8,12,15,20],"min_samples_split":[15, 18]})
confusion_matrix(yts,gs.fit(xtr,ytr).predict(xts))
accuracy_score(yts,gs.fit(xtr,ytr).predict(xts))
```

```python
y=df["Position"]


dtclf=DecisionTreeClassifier()
x=df[["Age","Projects Completed","Productivity (%)","Satisfaction Rate (%)","Feedback Score","Salary"]]
xtr,xts,ytr,yts=train_test_split(x,y,random_state=0,test_size=0.2)
gs=GridSearchCV(dtclf,param_grid={"criterion":['entropy','gini'],"max_leaf_nodes":[8,12,15,20],"min_samples_split":[15, 18]})

accuracy_score(yts,gs.fit(xtr,ytr).predict(xts))
gs.best_params_
```

```python
pd.DataFrame(confusion_matrix(yts,gs.fit(xtr,ytr).predict(xts)),columns=df["Position"].unique(),index=df["Position"].unique())
```
