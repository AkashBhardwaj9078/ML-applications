```python
import pandas as pd
import numpy as np
```

```python
df=pd.read_csv("/content/Advertising_Data (1).csv")
```

```python
from sklearn.model_selection import train_test_split,GridSearchCV,KFold,cross_val_score
from sklearn.tree import DecisionTreeRegressor
from sklearn.linear_model import Ridge,Lasso,LogisticRegression
from sklearn.ensemble import RandomForestRegressor
from sklearn.metrics import r2_score,classification_report
```

```python
x=df.iloc[:,:-1]
y=df.iloc[:,-1]
```

```python
xtr,xts,ytr,yts=train_test_split(x,y,random_state=0,test_size=0.4)
```

```python
mod1=DecisionTreeRegressor()
mod1.get_params()
params={'max_depth':[4,8,12],"max_features":[5,10,30],'ccp_alpha': [0.5,2.0,4.0,8.0]}

```

```python
np.mean(cross_val_score(mod1,x,y,cv=KFold(5)))
```

```python
grid=GridSearchCV(mod1,param_grid=params,cv=KFold(10),scoring="neg_mean_squared_error")
```

```python
pred=grid.fit(xtr,ytr).predict(xts)
r2_score(pred,yts)
```

```python
print(grid.best_params_)
print(grid.best_score_)
```

```python
from sklearn.neighbors import KNeighborsRegressor
from sklearn.svm import SVC
```

```python
reg=KNeighborsRegressor()
reg.get_params()
```

```python
params={"leaf_size":[12,24,36],"n_jobs":[6,12,15,20],"p":[1,2],"weights":['uniform',"distance"]}
```

```python
gridreg=GridSearchCV(reg,param_grid=params,cv=KFold(10))
pred=gridreg.fit(xtr,ytr).predict(xts)
r2_score(pred,yts)
```
