```python
df['Age'].fillna(df['Age'].mean(),inplace=True)
```

```python
df.isnull().sum()
```

```python
x=df[['Survived','Pclass','Age','SibSp','Parch']]
```

```python
x
```

```python
y=df['Fare']
```

```python
from sklearn.ensemble import RandomForestRegressor
from sklearn.tree import DecisionTreeRegressor
```

```python
from sklearn.model_selection import train_test_split
xtr,xts,ytr,yts=train_test_split(x,y)
```

```python
print("train_inputdata_shape",xtr.shape)
print('test_inputdata_shape',xts.shape)
print("train_outputdata_shape",ytr.shape)
print('test_outputdata_shape',yts.shape)
```

```python
print(xtr)
print(ytr)
```

```python
print(xts)
print(yts)
```

```python
rf_reg=RandomForestRegressor()
rf_reg.fit(xtr,ytr)
predrf=rf_reg.predict(xts)
predrf
```

```python
from sklearn.metrics import r2_score
print("r2_score of randomforest regressor",r2_score(predrf,yts))
```

```python
sns.regplot(x=yts,y=predrf)
plt.ylabel('predrf')
plt.title("Random forest prediction")
```

```python
dt_reg=DecisionTreeRegressor()
dt_reg.fit(xtr,ytr)
preddt=dt_reg.predict(xts)
preddt
```

```python
print("r2_score of Decisiontree regressor",r2_score(preddt,yts))
```

```python

sns.regplot(x=yts,y=preddt)
plt.ylabel('preddt')
plt.title("Decision Tree prediction")
```
