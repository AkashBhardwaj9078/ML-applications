```python
from sklearn.preprocessing import StandardScaler,MinMaxScaler
from sklearn.model_selection import train_test_split

from sklearn.linear_model import LogisticRegression
from sklearn.svm import SVC
```

```python
x=df.select_dtypes('int64')
y=df.iloc[:,-1]
```

```python
x_scaled=StandardScaler().fit_transform(x)
x_mnmx=MinMaxScaler().fit_transform(x)
```

```python
xtrscaled,xtsscaled,ytr,yts=train_test_split(x_scaled,y,random_state=0,test_size=0.4)

print('training data.shape',xtr.shape)
print('testing data.shape',xts.shape)
```

```python
model_svc=SVC()
model_svc.fit(xtrscaled,ytr)
pred_scaled=model_svc.predict(xtsscaled)

pred_scaled
```

```python
from sklearn.metrics import classification_report,confusion_matrix,accuracy_score
```

```python
print("SVM accuracy_score of StandardScaled Values",accuracy_score(pred_scaled,yts))
print("SVM classification report of StandardScaled Values\n",classification_report(pred_scaled,yts))
print("SVM confusion Matrix of StandardScaled Values\n",confusion_matrix(pred_scaled,yts))
```

```python
xtrmnmx,xtsmnmx,ytr,yts=train_test_split(x_mnmx,y,random_state=0,test_size=0.4)

model_svc.fit(xtrmnmx,ytr)
pred_mnmx=model_svc.predict(xtsmnmx)

pred_mnmx
```

```python
print("accuracy_score of MinMaxScaled Values",accuracy_score(pred_mnmx,yts))
print("classification report\n",classification_report(pred_mnmx,yts))
print("confusion Matrix\n",confusion_matrix(pred_mnmx,yts))
```

```python
model_reg=LogisticRegression()
model_reg.fit(xtrmnmx,ytr)
pred_reg_mnmx=model_reg.predict(xtsmnmx)
print("Logisticregression accuracy_score of MinMaxScaled Values",accuracy_score(pred_reg_mnmx,yts))
print("Logisticregression classification report\n",classification_report(pred_reg_mnmx,yts))
print("Logisticregression confusion Matrix\n",confusion_matrix(pred_reg_mnmx,yts))
```
