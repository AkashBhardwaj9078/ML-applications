```python
df.Ram.value_counts()
df.Ram=df.Ram.str.replace("GB"," ").astype(int)
```

```python
df.ROM.value_counts()
df.ROM=df.ROM.str.replace("GB"," ")
df.ROM=df.ROM.str.replace("TB","000")
df.ROM=df.ROM.astype(int)
```

```python
df=df.iloc[:,2:]
```

```python
from sklearn.preprocessing import LabelEncoder
lr1=LabelEncoder()
df.Ram_type=lr1.fit_transform(df.Ram_type)
lr2=LabelEncoder()
df.ROM_type=lr2.fit_transform(df.ROM_type)
```

```python
df
```

```python
x=df.drop(["name","processor","CPU","GPU","price","OS","brand"],axis=1)
y=df.price
```

```python
print(x)
print(y)
```

```python
from sklearn.metrics import r2_score,mean_squared_error,classification_report
from sklearn.ensemble import RandomForestRegressor
from sklearn.model_selection import KFold,GridSearchCV,train_test_split
import numpy as np
```

```python
rf=RandomForestRegressor()
params={"min_samples_split":[4,7,12],"max_features":["sqrt","log2","None"],"criterion":["squared_error","absolute_error","friedman_mse","poisson"]}

gr=GridSearchCV(rf,param_grid=params,cv=KFold(5),scoring="neg_mean_squared_error")
```

```python
xtr,xts,ytr,yts=train_test_split(x,y,random_state=42,test_size=0.3)
```

```python
rf.fit(xtr,ytr)
pred=rf.predict(xts)
print("r2_score :",r2_score(pred,yts))
print("mean_squared_error :",np.sqrt(mean_squared_error(pred,yts)))
```

```python
sns.scatterplot(x=yts,y=yts-pred)
```

```python
feature=pd.DataFrame(rf.feature_importances_,columns=["feature_Imp"])
feature["feature"]=x.columns

sorted_feature=feature.sort_values("feature_Imp",ascending=False)
sorted_feature
```
